```python
import cv2
from mtcnn import MTCNN
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import tensorflow as tf
from tensorflow.keras.applications import ResNet50, DenseNet121
from tensorflow.keras.layers import MultiHeadAttention, Flatten, Dense, Input, Concatenate
from tensorflow.keras.models import Model

# Initialize the MTCNN detector
detector = MTCNN()

# Paths to training and testing data
train_dir = '/content/train'
test_dir = '/content/test'

# Create ImageDataGenerators for loading and augmenting images
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)

test_datagen = ImageDataGenerator(rescale=1./255)

# Load training data
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(128, 128),
    color_mode='grayscale',
    class_mode='categorical',
    batch_size=64
)

# Load testing data
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(128, 128),
    color_mode='grayscale',
    class_mode='categorical',
    batch_size=64
)

# Example code to visualize a batch of images from the training set
x_batch, y_batch = next(train_generator)
plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(x_batch[i].reshape(128, 128), cmap='gray')
    plt.axis('off')
plt.show()

# Load an image using OpenCV (adjust the path to your image file)
image_path = '/content/sample_data/Screenshot (311).png'
image = cv2.imread(image_path)

if image is None:
    print(f"Error: Could not load image from {image_path}. Please check the path and file permissions.")
else:

    # Resize the image to a fixed size (e.g., 800x600)
    image_resized = cv2.resize(image, (800, 600))

    # Apply data augmentation (e.g., horizontal flip, random rotation)
    if np.random.rand() > 0.5:
        image_resized = cv2.flip(image_resized, 1)  # Horizontal flip

    rows, cols = image_resized.shape[:2]
    rotation_matrix = cv2.getRotationMatrix2D((cols / 2, rows / 2), 0, 1)
    image_resized = cv2.warpAffine(image_resized, rotation_matrix, (cols, rows))

    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image_resized, cv2.COLOR_BGR2GRAY)

    # Normalize the image (scaling pixel values to range [0, 1])
    normalized_image = gray_image / 255.0

    # Convert the grayscale image back to RGB for MTCNN (to display colored bounding boxes)
    image_rgb = cv2.cvtColor((normalized_image * 255).astype('uint8'), cv2.COLOR_GRAY2RGB)

    # Perform face detection
    faces = detector.detect_faces(image_rgb)

    # Draw bounding boxes and keypoints on the image
    for face in faces:
        # Get the bounding box
        x, y, width, height = face['box']
        cv2.rectangle(image_rgb, (x, y), (x + width, y + height), (0, 255, 0), 2)

        # Get the keypoints (eyes, nose, mouth)
        keypoints = face['keypoints']
        for point in keypoints.values():
            cv2.circle(image_rgb, point, 2, (255, 0, 0), -1)

    # Display the image with detected faces and keypoints
    plt.imshow(image_rgb)
    plt.axis('off')
    plt.show()
```

```python
# Global feature extraction using ResNet-50
input_tensor = Input(shape=(128, 128, 1))
x = tf.keras.layers.Conv2D(3, (3, 3), padding='same')(input_tensor)  # Convert grayscale to RGB

# Instantiate ResNet50 with 'imagenet' weights and 3 input channels
resnet_model = ResNet50(include_top=False, weights='imagenet', input_shape=(128, 128, 3), pooling='avg')

# Feed the converted input (x) to the ResNet50 model
global_features = resnet_model(x)
global_features = Flatten()(global_features)

# Local feature extraction using DenseNet
# Instantiate DenseNet121 with 'imagenet' weights and 3 input channels
densenet_model = DenseNet121(include_top=False, weights='imagenet', input_shape=(128, 128, 3), pooling='avg')

# Feed the converted input (x) to the DenseNet121 model
local_features = densenet_model(x)
local_features = Flatten()(local_features)

# Project the features to the same dimension using Dense layers
# You can adjust the output dimension (e.g., 512) as needed
global_features = Dense(512, activation='relu')(global_features)  # Project global features
local_features = Dense(512, activation='relu')(local_features)    # Project local features

# Reshape features for multi-head attention using tf.keras.layers.Reshape
# (batch_size, seq_length=1, feature_dim)
global_features_reshaped = tf.keras.layers.Reshape((1, global_features.shape[-1]))(global_features)  # Use Keras Reshape
local_features_reshaped = tf.keras.layers.Reshape((1, local_features.shape[-1]))(local_features)  # Use Keras Reshape

# Multi-head attention layer
attention_layer = MultiHeadAttention(num_heads=4, key_dim=128)
attention_output = attention_layer(query=global_features_reshaped, value=local_features_reshaped)

# Flatten the attention output for concatenation
attention_output_flattened = Flatten()(attention_output)

# Merging the features
merged_features = Concatenate()([global_features, local_features, attention_output_flattened])

# Classification layer
output = Dense(train_generator.num_classes, activation='softmax')(merged_features)

# Final model
model = Model(inputs=input_tensor, outputs=output)

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Summary of the model
model.summary()
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder # Import LabelEncoder


# Assuming `model` is the trained emotion detection model from the previous code
# and `frames` is a batch of images (e.g., a sequence of frames for a video or set of images for an individual).

def predict_and_aggregate(model, frames, encoder):
    """
    Predict emotions for each frame in a sequence and aggregate the result.

    Parameters:
    - model: Trained Keras model for emotion detection.
    - frames: NumPy array of shape (num_frames, 48, 48, 1), a sequence of grayscale frames.

    Returns:
    - aggregated_emotion: Final aggregated emotion label (string or class index).
    - aggregated_probs: Mean probabilities for each emotion class.
    """
    # Step 1: Predict probabilities for each frame
    predictions = model.predict(frames)  # shape (num_frames, num_classes)

    # Step 2: Average probabilities across frames
    aggregated_probs = np.mean(predictions, axis=0)  # shape (num_classes,)

    # Step 3: Find the emotion with the highest average probability
    aggregated_emotion_index = np.argmax(aggregated_probs)
    aggregated_emotion_label = encoder.classes_[aggregated_emotion_index]  # Map to label

    return aggregated_emotion_label, aggregated_probs

# Example usage:
# Assume `sequence_frames` is a numpy array with multiple frames for a single session or person
# shape: (num_frames, 48, 48, 1)
# Get the first 10 images from the test generator
test_images, _ = next(test_generator)

sequence_frames = np.expand_dims(test_images, axis=-1)  # Use test_images instead of test

# Initialize LabelEncoder with class labels from the training generator
encoder = LabelEncoder()
encoder.fit(train_generator.classes)

aggregated_emotion, aggregated_probs = predict_and_aggregate(model, sequence_frames, encoder)  # Pass encoder

print("Aggregated Emotion:", aggregated_emotion)
print("Aggregated Probabilities:", aggregated_probs)
```
